### Governing equations

The governing equations are 
$$
a^{t^n} = - \omega_0^2 x^{t^n},
$$

$$
x^{t^{n+1}} = x^{t^n} + v^{t^n} \times \Delta t,
$$
and 
$$
v^{t^{n+1}} = v^{t^n} + a^{t^n} \times \Delta t.
$$

### Initial Conditions

At time $t=0$, position $x=1$ and velocity $v=0$.\
Set $A=k=m=1$, and $\omega_0=\sqrt{k/m}=1$ as well.

## Exercie 1: 
Use a small time step $\Delta t=0.01$ and solve for the solution at $t=20$.

```python
# import required libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
#
# This is a simple example of how to solve a simple harmonic oscillator using the Euler method
#

# Step 1: set up the parameters of the problem
A=1
k=1
m=1
omega0 = np.sqrt(k/m)
dt = 0.01

# Step 2: set up the time and solution arrays
dt = 0.01  
t_final = 20  
n_steps = int(t_final / dt) 
time = np.arange(0, t_final + dt, dt)  


x_array = np.zeros(len(time))  
v_array = np.zeros(len(time))



# Step 3: set up the initial conditions
x_array[0] = 1 
v_array[0] = 0 




# Step 4: solve the difference equation using the Euler method
omega0 = 1
for n in range(len(time) - 1):
    a_n = -omega0**2 * x_array[n]
    x_array[n + 1] = x_array[n] + v_array[n] * dt
    v_array[n + 1] = v_array[n] + a_n * dt



```

```python
# Step 5: plot the solution

#TODO

plt.figure(figsize=(10, 6))
plt.plot(time, x_array, label='Numerical Solution (Euler Method)')
plt.title('Position vs. Time')
plt.xlabel('Time')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.show()



```

We could verify our numerical solution be comparing it with the analytical solutions.
The analytical solutions are:

$$
x = A \cos(\omega_0 t + \phi),
$$
and

$$
v = -A \omega_0 \sin(\omega_0 t + \delta).

```python
# Step 6: evaluate the analytical solution and plot it

# TODO
x_analytical = np.cos(omega0 * time)
plt.figure(figsize=(10, 6))
plt.plot(time, x_array, label='Numerical Solution (Euler Method)', linestyle='-', color='blue')
plt.plot(time, x_analytical, label='Analytical Solution', linestyle='--', color='red')
plt.title('Position vs. Time')
plt.xlabel('Time')
plt.ylabel('Position')
plt.grid(True)
plt.legend()
plt.show()






```

Another way to chekc the accuray of our numerical solution is to check the energy conservation and the phase-sapce diagram.

```python
# Step 7: evaluate the energy (error) of the system

# TODO 

kinetic_energy = 0.5 * v_array**2
potential_energy = 0.5 * x_array**2
total_energy = kinetic_energy + potential_energy

plt.figure(figsize=(10, 6))
plt.plot(time, total_energy, label='Total Energy', linestyle='-', color='green')
plt.title('Total Energy vs. Time')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.grid(True)
plt.legend()
plt.show()



```

```python
# Step 8: evaluate the phase space

# TODO
plt.figure(figsize=(10, 6))
plt.plot(x_array, v_array, label='Phase Space Trajectory', color='purple')
plt.title('Phase Space: Velocity vs. Position')
plt.xlabel('Position')
plt.ylabel('Velocity')
plt.grid(True)
plt.legend()
plt.show()






```
